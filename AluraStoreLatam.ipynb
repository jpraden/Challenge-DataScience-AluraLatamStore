# **1. Análisis de facturación**

Trello: https://trello.com/c/6CSot4po/20-ingreso-total-por-cada-tienda

**Ingreso total por cada tienda**

**Descripción**
En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

**Extra**
https://www.aluracursos.com/blog/listas-de-python-operaciones-basicas

```python
#1. Análisis de facturación: Ingresto Total por cada TIENDA
# Cálculo del ingreso total por tienda
ingresoTotal_tienda1 = tienda1['Precio'].astype(float).sum()
ingresoTotal_tienda2 = tienda2['Precio'].astype(float).sum()
ingresoTotal_tienda3 = tienda3['Precio'].astype(float).sum()
ingresoTotal_tienda4 = tienda4['Precio'].astype(float).sum()
```

```python
# Factutación total de todas las tiendas
facturacion_total = ingresoTotal_tienda1 + ingresoTotal_tienda2 + ingresoTotal_tienda3 + ingresoTotal_tienda4
print(f"\nFacturación ingreso total  - Todas las Tiendas: $ {facturacion_total:,.2f}")

# Mostrando resultados
print(f"Ingreso total Tienda1: $ {ingresoTotal_tienda1:,.2f}")
print(f"Ingreso total Tienda2: $ {ingresoTotal_tienda2:,.2f}")
print(f"Ingreso total Tienda3: $ {ingresoTotal_tienda3:,.2f}")
print(f"Ingreso total Tienda4: $ {ingresoTotal_tienda4:,.2f}")
```

**RESULTADOS** - Ingresos totales por cada tienda **(FACTURACION)**

  -Ingreso total Tienda1: **$ 1,150,880,400.00**
