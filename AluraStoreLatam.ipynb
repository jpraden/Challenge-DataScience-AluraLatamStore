# **2. Ventas por Categoría**
En este debes calcular la **CANTIDAD de productos vendidos por categoría en cada tienda**. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.



**Definimos variables para:**

    - cntpvc_tienda1 : almacenamos la cantidad de ventas por categoría de productos vendidos para la tienda1

    - cntpvc_tienda2 : Almacenamos la cantidad de ventas por categoría de productos vendidos para la tienda2

    - cntpvc_tienda4 : Almacenamos la cantidad de ventas por categoría de productos vendidos para la tienda4

2. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda:
    - se crea una función **"top5_cntventas_por_categoria"** para dar solución a la agrupación de la **cantidad** de productos vendidos para mostrar las categorías más populares por casa tienda    

```python
tienda1.columns
```

**1.- Cálculamos CANTIDAD de productos vendidos por categoría en cada tienda**

```python
# cntvpc_tienda#: Cantidad de Ventas por Categoria de productos vendidos de cada tienda
cntpvc_tienda1 = tienda1['Categoría del Producto'].value_counts()
cntpvc_tienda2 = tienda2['Categoría del Producto'].value_counts()
cntpvc_tienda3 = tienda3['Categoría del Producto'].value_counts()
cntpvc_tienda4 = tienda4['Categoría del Producto'].value_counts()
```

**2.-La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo**

```python
#Mostramos Cant.Ventas por Producto Vendido por categorías de cada tienda
print(f"Tienda1 - Cant.Ventas por Producto Vendido: {cntpvc_tienda1}\n")
print(f"Tienda2 - Cant.Ventas por Producto Vendido: {cntpvc_tienda2}\n")
print(f"Tienda3 - Cant.Ventas por Producto Vendido: {cntpvc_tienda3}\n")
print(f"Tienda4 - Cant.Ventas por Producto Vendido: {cntpvc_tienda4}\n")
```

**Mostramos Resultados TOP 5 - Categorías más populares de CANTIDAD de productos vendidos de cada tienda.**

```python
#top5_cntventas_por_categoria : top5 de CANTIDAD de productos vendidos de las categorías más populares de cada tienda
def top5_cntventas_por_categoria(tienda, numero):
    ventas = tienda['Categoría del Producto'].value_counts().sort_values(ascending=False).head(5)
    print(f"\n***** TOP 5 - Categorías más populares de CANTIDAD de productos vendidos - Tienda {numero}: *****")
    print(ventas)

# Mostrar resultados para cada tienda
top5_cntventas_por_categoria(tienda1, 1)
top5_cntventas_por_categoria(tienda2, 2)
top5_cntventas_por_categoria(tienda3, 3)
top5_cntventas_por_categoria(tienda4, 4)
```

**Otra análisis, dando una mirada por Ingresos Totales por categorías de productos vendidos**

Se muestra los RESULTADOS agrupados, **TOP 5** en categorías más populares de productos vendidos **(x PRECIO)** de cada tienda, para tener otra mirada y ver si existe alguna DIFERENCIA en el ordenamiento, respecto a la **CANTIDAD** de ventas por categorías más populares.


```python
#top5_precioventas_categoria : top5 de PRECIO de productos vendidos de las categorías más populares de cada tienda
# Función para calcular y mostrar las ventas por categoría de una tienda - TOP5
def top5_precioventas_categoria(tienda, numero):
    precioventas_categoria = tienda.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False).head()
    print(f"\n***** TOP 5 - Categoría de Productos más vendidos (PRECIO) - Tienda{numero}: *****")
    print(precioventas_categoria)

# Mostrar resultados para cada tienda
top5_precioventas_categoria(tienda1, 1)
top5_precioventas_categoria(tienda2, 2)
top5_precioventas_categoria(tienda3, 3)
top5_precioventas_categoria(tienda4, 4)
```

**Lo que encontramos en la comparación**

Vemos que en el ordenamiento es distinto para todas las tiendas en los **TOP5** de las categorás más populares, comparando los resultados obtenidos por **CANTIDAD de ventas VS Ingresos totales (PRECIO)**.

- Si vemos los resultados de cada tienda, para las categorias más populares por **CANTIDAD de ventas**. Se aprecia que tenemos un ordenamiento igual para las primeras 3 categorías en todas las tiendas, No obstante, son distintos el ordemaniento en la **"Tienda4"** en las posisiones 4 y 5,  respecto a las misma posisiones de ordenamiento con las otras tiendas.

- Ahora, si vemos los resultados de cada tienda, para las categorias más populares por **Ingresos Totales de ventas (PRECIO)**. Se aprecia que tenemos un ordenamiento identico para todas las tiendas y en comparación con el Top5 por "**CANTIDAD de ventas**", se registra un cambio en la distribución en las categorias más populares.


# **3. Valoración media por tienda**
En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
#Agrupamos en un diccionario la calificación promedio de satisfacción de los clientes de cada tienda
calificaciones_promedio_tiendas = {}
calificaciones_promedio_tiendas['Tienda1'] = tienda1['Calificación'].mean()
calificaciones_promedio_tiendas['Tienda2'] = tienda2['Calificación'].mean()
calificaciones_promedio_tiendas['Tienda3'] = tienda3['Calificación'].mean()
calificaciones_promedio_tiendas['Tienda4'] = tienda4['Calificación'].mean()

# Convertir a una Serie de pandas para facilitar el ordenamiento
calificaciones_series = pd.Series(calificaciones_promedio_tiendas)

# Ordenar de forma descendente
calificaciones_series = pd.Series(calificaciones_promedio_tiendas)

# Ordenar de forma descendente
calificaciones_ordenadas = calificaciones_series.sort_values(ascending=False)

print("\nClasificaciones promedio de satisfacción de clientes por tienda (orden descendente):")
for tienda, promedio in calificaciones_ordenadas.items():
    print(f"{tienda}: {promedio:.3f}")
```

**Cálculo de Ingresos y Ventas agrupadas**

Si tenemos una columna "Cantidad" en los CSV, debemos **USAR** DataFrame Unificado --> tiendas_all [Cantidad] * tiendas_all['Precio']

Se asume y fue confirmada que cada registro es una venta, donde el 'Precio' representa el total de la transacción unitaria.

**Ahora vamos a generar DataFrame de resumen de ventas**
- "ventas_ingresos_totales"
- "ventas_cantidad_por_producto"



```python
#Cálculo de Ingresos de ventas Agrupadas
ventas_ingresos_totales = tiendas_all.groupby(['tienda_id', 'Producto'])['Precio'].sum().reset_index(name='Ingresos_Totales')
#Cálculo de Cantidad de ventas Agrupadas
ventas_cantidad_por_producto = tiendas_all.groupby(['tienda_id', 'Producto']).size().reset_index(name='IngreCantidad')
```

**Unimos los DataFrames:**
- ventas_ingresos_totales
- ventas_cantidad_por_producto

Con Merge, con el fin de generar un nuevo DataFrame de RESUMEN para mostrar Productos más y menos vendidos por Ingresos Totales (**foco ordenamiento**) y ver la cantidades de ventas de los productos.

```python
# Unir ambos DataFrames
ventas_merged = pd.merge(
    ventas_ingresos_totales.rename(columns={'tienda_id': 'Tienda'}),
    ventas_cantidad_por_producto.rename(columns={'tienda_id': 'Tienda'}),
    on=['Tienda', 'Producto'],
    how='left'
)
ventas_merged.info()
```

**Ahora que tenemos la información de resumen consolidada:**

En este paso, **"4. Productos más vendidos y menos vendidos"**, con foco a los productos que generaron mayores y menores ingresos totales, incluyendo la cantidad de ventas y así visualizar los resultados de los productos destacaron en ventas en cada tienda.

```python
print("****************************************************************************************************************")
print("✓✓ TOP3 - Productos MÁS y MENOS Ingresos Totales generaron (con Cantidad de Ventas asociada) ✓✓")
print("****************************************************************************************************************")

for tienda in sorted(ventas_merged['Tienda'].unique()):
    df_tienda = ventas_merged[ventas_merged['Tienda'] == tienda]

    # Top 3 productos con MAS Ingresos Totales generados
    top_ingresos = df_tienda.sort_values(by='Ingresos_Totales', ascending=False).head(3)
    print(f"\n\033[1m***** Tienda {tienda} - Productos que MÁS INGRESOS TOTALES generaron: *****\033[1m")
    #display(top_ingresos[['Producto', 'Ingresos_Totales', 'IngreCantidad']].reset_index(drop=True))

    # Formatear la columna con separador de miles
    df_top = top_ingresos[['Producto', 'Ingresos_Totales', 'IngreCantidad']].reset_index(drop=True).copy()
    df_top['Ingresos_Totales'] = df_top['Ingresos_Totales'].apply(
        lambda x: f"{x:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")
    )

    display(df_top)

    # Top 3 productos con MENOS Ingresos Totales generados
    bottom_ingresos = df_tienda.sort_values(by='Ingresos_Totales', ascending=True).head(3)
    print(f"\n\033[1m***** Tienda {tienda} - Productos que MENOS INGRESOS TOTALES generaron: *****\033[1m")
    #display(bottom_ingresos[['Producto', 'Ingresos_Totales', 'IngreCantidad']].reset_index(drop=True))

    # Formatear la columna con separador de miles
    df_bottom = bottom_ingresos[['Producto', 'Ingresos_Totales', 'IngreCantidad']].reset_index(drop=True).copy()
    df_bottom['Ingresos_Totales'] = df_bottom['Ingresos_Totales'].apply(
        lambda x: f"{x:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")
    )

    display(df_bottom)
```

**Grafico #1 - Participación de cada tienda (Ingresos Totales por Tienda)**

```python
# Agrupamos ingresos totales por tienda
ingresos_totales = tiendas_all.groupby('tienda_id')['Precio'].sum()
ingresos_totales.index = [f"Tienda{i}" for i in ingresos_totales.index]

# Gráfico de pastel con etiquetas externas
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(
    ingresos_totales,
    labels=ingresos_totales.index,
    autopct=lambda p: f'{p:.1f}%',   # formato porcentual
    colors=['#021526', '#2513F0', '#1375EC', '#E51005'],
    startangle=90,
    pctdistance=0.8,   # distancia del porcentaje respecto al centro
    labeldistance=1.1  # distancia de las etiquetas fuera del pastel
)

# Ajustes de estilo
for autotext in autotexts:
    autotext.set_color('white')       # porcentajes en blanco
    autotext.set_fontsize(12)
    autotext.set_weight("bold")

for text in texts:
    text.set_fontsize(12)
    text.set_weight("bold")

plt.title('Gráfico #1 \nParticipación de cada Tienda en los Ingresos Totales', fontsize=15)
plt.show()
```

 **El gráfico #1**, evidencia que existe una brecha significativa en la participación de ingresos entre tiendas, destacando a la **Tienda4** nos entrega el menor aporte. **Grafico #2 - Ingresos Totales por Tienda (Tienda 4 = Menor Ingresos)**

```python
# Agrupamos ingresos totales por tienda
ingresos_totales = tiendas_all.groupby('tienda_id')['Precio'].sum()
ingresos_totales.index = [f"Tienda{i}" for i in ingresos_totales.index]

#Definimos los colores Rojo para la tienda4 de menor rendimiento, respecto a los ingresos
colores_bar = ['#021526', '#2513F0', '#1375EC', '#E51005']

# Definición de la figura
plt.figure(figsize=(10, 6))
ax = ingresos_totales.plot(kind='bar', color=colores_bar)

# Personalización
plt.title('Gráfico #2 \n Ingresos Totales por Tienda (Tienda 4 = Menor Ingresos)', fontsize=15)
plt.xlabel('Tienda')
plt.ylabel('Ingresos Totales (USD)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Eliminar líneas del cuadro superior y derecho
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Añadir valores encima de cada barra
for p, label in zip(ax.patches, ingresos_totales.index):
    valor = f'{p.get_height():,.2f}'
    x = p.get_x() + p.get_width() / 2.
    y = p.get_height()

    # anotación del valor
    ax.annotate(valor,
                (x, y),
                ha='center', va='center',
                xytext=(0, 10),
                textcoords='offset points')

    # Si es la Tienda4, dibujamos un óvalo semitransparente alrededor del texto
    if label == "Tienda4":
        ellipse = patches.Ellipse(
            (x, y + 15),   # centro en la posición del texto
            width=0.85,     # ancho del óvalo (ajusta según tamaño del número)
            height=0.15 * y,   # alto relativo al valor
            edgecolor='red', facecolor='yellow', alpha=0.7, linewidth=3
        )
        ax.add_patch(ellipse)

plt.show()
```

**El gráfico #2**, nos entrega el análisis de facturación y nos muestra que las tiendas tienen un desempeño desigual en términos de ingresos totales.

En particular, la **Tienda4** destaca por ser la de **MENOR** rendimiento (el cual se resalta su valor con un óvalo visualmente), lo que sugiere que podría requerir una revisión de estrategias comerciales, promociones o surtido de productos.

Las demás tiendas presentan ingresos significativamente superiores, lo que evidencia un potencial de buenas prácticas que podrían replicarse en la **Tienda4** con menor facturación.

```python
# 1. First, ensure 'Fecha de Compra' is a datetime object tiendas_all
tiendas_all['Fecha de Compra'] = pd.to_datetime(tiendas_all['Fecha de Compra'], dayfirst=True)

# Calculate monthly sales for each store
# Group by year-month and Tienda_ID, then sum the 'Precio'
ventas_mensuales = tiendas_all.groupby(
    [tiendas_all['Fecha de Compra'].dt.to_period('M'), 'tienda_id']
)['Precio'].sum().reset_index()

# Rename columns for clarity and to match the expected 'Tienda' column name in the original code
ventas_mensuales.columns = ['Fecha', 'Tienda', 'Ingresos_Mensuales']

# Convert the PeriodIndex to Timestamp for plotting
ventas_mensuales['Fecha'] = ventas_mensuales['Fecha'].dt.to_timestamp()
ventas_mensuales = ventas_mensuales.set_index('Fecha')

# 2. Definimos la paleta de colores para destacar la Tienda 4
# Mapeamos los colores a las tiendas: Tienda 4 debe ser roja
color_map = {'Tienda 1': '#1f77b4', 'Tienda 2': '#ff7f0e', 'Tienda 3': '#2ca02c', 'Tienda 4': 'red'}

plt.figure(figsize=(12, 6))

for tienda_id in ventas_mensuales['Tienda'].unique():
    # Format tienda_id to match dictionary keys (e.g., 1 -> 'Tienda 1')
    tienda_label = f'Tienda {tienda_id}'
    data_tienda = ventas_mensuales[ventas_mensuales['Tienda'] == tienda_id].copy()
    data_tienda['Media_Movil_6M'] = data_tienda['Ingresos_Mensuales'].rolling(window=6).mean()

    # Graficamos, usando el color específico para la tienda
    sns.lineplot(data=data_tienda, x=data_tienda.index, y='Media_Movil_6M',
                 label=tienda_label, marker='', color=color_map[tienda_label])

plt.title('Tendencia Suavizada de Ventas Mensuales (Media Móvil 6M) - Destacando Tienda 4', fontsize=15)
plt.ylabel('Ingreso Promedio Suavizado (USD)')
plt.xlabel('Fecha')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend(title='Tienda')
plt.tight_layout()
plt.show()
```
